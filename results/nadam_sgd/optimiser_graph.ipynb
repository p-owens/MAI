```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
nadam_pd = pd.read_csv(os.getcwd() + "/ch19_training_log_nadam.csv")
sgd_pd = pd.read_csv(os.getcwd() + "/ch19_training_log_sgd.csv")
nadam = nadam_pd.to_numpy()
sgd = sgd_pd.to_numpy()
file_ext = 'pdf'
```

```python
nadam_pd
```

```python
fig, ax = plt.subplots(1, figsize = (12,8))

plt.plot(nadam[:,0], nadam[:,1], label='Nadam')

plt.plot(sgd[:,0], sgd[:,1], label='SGD')



title = 'SGD v Nadam, Ch 19 Training Log'
ax.set_title(title)
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss: Mean Squared Error')


ax.set_ylim([0, 2.6])
plt.legend(loc="best")
plt.show()

fig.savefig('images/{0}.{1}'.format(title,file_ext))
```

```python
fig = plt.figure(figsize = (10,6))
a1 = fig.add_axes([0,0,1,1])

a1.plot(nadam[:,0], nadam[:,1], label='Mean Squared Error', color='tab:blue')

a2 = a1.twinx()
a2.plot(nadam[:,0], nadam[:,2], label='Mean Absolute Percentage Error', color='tab:orange')

a1.set_ylim([0, 2.7])
a2.set_ylim([0, 5.6e7])


title = 'Loss Function Comparison, Mean Squared Error v Mean Absolute Percentage Error'
a1.set_title(title)
a1.set_xlabel('Epochs')
a1.set_ylabel('Loss: Mean Squared Error')
a2.set_ylabel('Loss: Mean Absolute Percentage Error')

fig.legend(labels = ('Mean Squared Error','Mean Absolute Percentage Error'),loc='upper right', bbox_to_anchor=(0.99, 0.99))
plt.show()

fig.savefig('images/{0}.{1}'.format(title,file_ext))
```

```python
fig2, ax2 = plt.subplots(1, figsize = (10,6))

plt.plot(nadam[:,0], nadam[:,1], label='Training Loss')

plt.plot(nadam[:,0], nadam[:,4], label='Validation Loss')


ax2.set_ylim([0, 2.65])

title = 'Channel 19 Training Log'
ax2.set_title(title)
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss: Mean Squared Error')

plt.legend(loc="best")
plt.show()

fig2.savefig('images/{0}.{1}'.format(title,file_ext))
```

```python
fig2, ax2 = plt.subplots(1, figsize = (10,6))

plt.plot(nadam[100:,0], nadam[100:,1], label='Training Loss')

plt.plot(nadam[100:,0], nadam[100:,4], label='Validation Loss')


ax2.set_ylim([0, 0.33])

title = 'Channel 19 Training Log'
ax2.set_title(title)
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss: Mean Squared Error')

plt.legend(loc="best")
plt.show()

fig2.savefig('images/{0}.{1}'.format(title+'inset',file_ext))
```
